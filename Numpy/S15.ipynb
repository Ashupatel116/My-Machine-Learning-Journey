## what is Pandas

## Pandas Series 
A pandas Series like a column in a table . It is a 1-D array holding data of any type 

## Importing pandas 

```python

import numpy as np 
import pandas as pd 
```

## Series from list 

```python
# String 
country = ['India','pakistan','USA','Nepal','Shrilanka']

pd.Series(country)
```

```python
# integer
runs=[13,24,35,46,57,68]
pd.Series(runs)
```

```python
# custom index 
marks = [89,85,99,100]
subjects = ['Maths','Enslish','Science','Hindi']
pd.Series(marks , index = subjects)
```

```python
# setting a name 
pd.Series(marks , index = subjects , name = 'Ashu Marks' )
```

## Sereis From dicti....

```python
marks = {
"Maths"     :   89,
"Enslish"   :   85,
"Science"   :   99,
"Hindi   "  :   100,
}

marks_series = pd.Series(marks,name='ASHU Marks')
marks_series 
```
