## API 
* application porgramming interface 

An API (Application Programming Interface) is a set of rules and protocols that allows different software systems to communicate. It acts as a bridge between applications, enabling them to exchange data and functionality without exposing internal code. For example, a weather app uses an API to fetch live weather data from a remote server instead of building its own weather system.

```python
import pandas as pd
import requests
import time

session = requests.Session()
all_movies = []

for i in range(1, 40):   # top_rated usually < 40 pages

    try:
        response = session.get(
            f"https://api.themoviedb.org/3/movie/top_rated?api_key=a58cf79c3e74d4d0df26fcb57031e303&page={i}",
            headers={"User-Agent": "Mozilla/5.0"},
            timeout=20
        )

        if response.status_code != 200:
            break

        data = response.json()

        if not data.get("results"):
            break

        temp_df = pd.DataFrame(data["results"])[
            ['id','title','overview','release_date',
             'popularity','vote_average','vote_count']
        ]

        all_movies.append(temp_df)

        print("Page done:", i)

        time.sleep(1.2)   # ⭐ increase delay

    except Exception as e:
        print("Stopped at page:", i)
        print(e)
        time.sleep(3)
        continue   # retry next page instead of breaking

df = pd.concat(all_movies, ignore_index=True)
print(df.shape)
```

```python
df.to_csv('movies.csv')
```
