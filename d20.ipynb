## Data 
Two types (1) Numerical , (2) Categorical

```python
import pandas as pd 
import seaborn as sns 
```

```python
df = pd.read_csv('d19/train.csv')
```

```python
df.head()
```

## Categorical Data 

### a. Countplot

```python
sns.countplot(x='Pclass', data=df)
#  df['Survived'].value_counts().plot(kind='bar')
```

### b. pie chart

```python
df['Survived'].value_counts().plot(kind='pie' , autopct = '%.2f')
```

## 2. Numerical Data 

### a. Histogram 

```python
import matplotlib.pyplot as plt 
```

```python
plt.hist(df['Age'] , bins=50)
```

### b. Distplot

```python
sns.displot(df["Age"]) 
```

### c. Boxplot

```python
sns.boxplot(df['Age'])
```
